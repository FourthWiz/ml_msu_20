## –ê–ª–≥–æ—Ä–∏—Ç–º—ã –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏

```python
#create data from normal distribution
x_dots = 300 #number of objects
num_clusters = 3 #number of clusters
size = x_dots // num_clusters #size of each cluster
locs = [-2, 0, -5]
scales = [0.6, 0.7, 0.3]

X = np.zeros((x_dots, 2))

np.random.seed(seed=SEED)
for i in range(num_clusters):
    X[i*size:i*size+size, 0] = np.random.normal(loc=locs[i], scale=scales[i], size=size)
    X[i*size:i*size+size, 1] = np.random.normal(loc=locs[i], scale=scales[i], size=size)

#plot data
plt.figure(figsize=(7, 7))
plt.scatter(X[:, 0], X[:, 1], c='blue');
```

–í —Å–∏—Ç—É–∞—Ü–∏–∏, –∫–æ–≥–¥–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–±—ä–µ–∫—Ç–æ–≤ –æ—á–µ–Ω—å –≤–µ–ª–∏–∫–æ, —á—Ç–æ–±—ã –Ω–µ –≤—ã—á–∏—Å–ª—è—Ç—å —Å—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ –ø–æ –≤—Å–µ–º –æ–±—ä–µ–∫—Ç–∞–º, –º–æ–∂–Ω–æ –Ω–∞ –∫–∞–∂–¥–æ–π –∏—Ç–µ—Ä–∞—Ü–∏–∏ —Å—á–∏—Ç–∞—Ç—å –ø–æ–ª–æ–∂–µ–Ω–∏—è —Ü–µ–Ω—Ç—Ä–æ–≤ –∫–ª–∞—Å—Å–æ–≤ –ø–æ –Ω–µ–∫–æ—Ç–æ—Ä–æ–π —Å–ª—É—á–∞–π–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ –æ–±—ä–µ–∫—Ç–æ–≤. –¢–∞–∫–æ–π –≤–∞—Ä–∏–∞–Ω—Ç –º–µ—Ç–æ–¥–∞ –ø–æ–∑–≤–æ–ª—è–µ—Ç –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –º–µ—Ç–æ–¥ K-Means –Ω–∞ –æ–≥—Ä–æ–º–Ω—ã—Ö –≤—ã–±–æ—Ä–∫–∞—Ö.

–°—Ö–æ–¥–∏–º–æ—Å—Ç—å –º–µ—Ç–æ–¥–∞ K-means —Å–∏–ª—å–Ω–æ –∑–∞–≤–∏—Å–∏—Ç –æ—Ç –≤—ã–±–æ—Ä–∞ –Ω–∞—á–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π —Ü–µ–Ω—Ç—Ä–æ–≤. –í —Å–ª—É—á–∞–µ –¥–≤—É—Ö –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ –º–æ–∂–Ω–æ –≤ –∫–∞—á–µ—Å—Ç–≤–µ —Ü–µ–Ω—Ç—Ä–æ–≤ –≤—ã–±—Ä–∞—Ç—å –¥–≤–µ —Å–∞–º—ã–µ —É–¥–∞–ª–µ–Ω–Ω—ã–µ –¥—Ä—É–≥ –æ—Ç –¥—Ä—É–≥–∞ —Ç–æ—á–∫–∏. –î–ª—è –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω–æ–≥–æ —á–∏—Å–ª–∞ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç –∞–ª–≥–æ—Ä–∏—Ç–º **k-means++**. –≠—Ç–æ –≤–∞—Ä–∏–∞—Ü–∏—è –º–µ—Ç–æ–¥–∞ K-means, –≤ –∫–æ—Ç–æ—Ä–æ–π –Ω–∞—á–∞–ª—å–Ω—ã–µ –ø—Ä–∏–±–ª–∏–∂–µ–Ω–∏—è –≤—ã–±–∏—Ä–∞—é—Ç—Å—è –Ω–µ —Å–ª—É—á–∞–π–Ω–æ, –∞ –Ω–µ–∫–æ—Ç–æ—Ä—ã–º –±–æ–ª–µ–µ –æ–ø—Ç–∏–º–∞–ª—å–Ω—ã–º –æ–±—Ä–∞–∑–æ–º: –ø–æ–ª–æ–∂–µ–Ω–∏–µ –ø–µ—Ä–≤–æ–≥–æ —Ü–µ–Ω—Ç—Ä–∞ –∫–ª–∞—Å—Å–∞ –≤—ã–±–∏—Ä–∞–µ—Ç—Å—è —Å–ª—É—á–∞–π–Ω–æ –∏–∑ —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –Ω–∞ –≤—ã–±–æ—Ä–∫–µ. –ù–∞ –∫–∞–∂–¥–æ–º —Å–ª–µ–¥—É—é—â–µ–º —à–∞–≥–µ —Å–Ω–∞—á–∞–ª–∞ –≤–≤–æ–¥–∏—Ç—Å—è —Ç–∞–∫–æ–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –Ω–∞ –≤—ã–±–æ—Ä–∫–µ, —á—Ç–æ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤—ã–±–æ—Ä–∞ —Ç–æ—á–∫–∏ –ø—Ä–æ–ø–æ—Ä—Ü–∏–æ–Ω–∞–ª—å–Ω–∞ –∫–≤–∞–¥—Ä–∞—Ç—É —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è –¥–æ –±–ª–∏–∂–∞–π—à–µ–≥–æ –¥–æ –Ω–µ–µ —Ü–µ–Ω—Ç—Ä–∞, –∞ –ø–æ–ª–æ–∂–µ–Ω–∏–µ –Ω–æ–≤–æ–≥–æ —Ü–µ–Ω—Ç—Ä–∞ –≤—ã–±–∏—Ä–∞–µ—Ç—Å—è –∏–∑ —ç—Ç–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è. 

### –ß—Ç–æ –æ–ø—Ç–∏–º–∏–∑–∏—Ä—É–µ—Ç k-means?

–°—Ä–µ–¥–Ω–µ–µ –≤–Ω—É—Ç—Ä–∏–∫–ª–∞—Å—Ç–µ—Ä–Ω–æ–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ: $F_0 = \frac{\sum_{i<j}[y_i=y_j]\rho(x_i; x_j)}{\sum_{i<j}[y_i=y_j]} \rightarrow min$

–ò–ª–∏, –∫–æ–≥–¥–∞ —Ü–µ–Ω—Ç—Ä—ã –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ –∏–∑–≤–µ—Å—Ç–Ω—ã: $\Phi_0 = \sum_{y \in Y}\frac{1}{|K_y|} \sum_{i:y_i=y}\rho^2{(x_i; \mu_y)} \rightarrow min$ - —Å—Ä–µ–¥–Ω–µ–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –¥–æ —Ü–µ–Ω—Ç—Ä–∞ –∫–ª–∞—Å—Ç–µ—Ä–∞

–í K-Means –∏—Ç–µ—Ä–∞—Ç–∏–≤–Ω–æ –º–∏–Ω–∏–º–∏–∑–∏—Ä—É–µ—Ç—Å—è —Å—Ä–µ–¥–Ω–∏–µ –≤–Ω—É—Ç—Ä–∏–∫–ª–∞—Å—Ç–µ—Ä–Ω—ã–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è: –æ–±—ä–µ–∫—Ç –ø—Ä–∏—Å–≤–∞–∏–≤–∞–µ—Ç—Å—è –∫ —Ç–æ–º—É –∫–ª–∞—Å—Ç–µ—Ä—É, —Ü–µ–Ω—Ç—Ä –∫–æ—Ç–æ—Ä–æ–≥–æ –±–ª–∏–∂–µ, –∞ —Ü–µ–Ω—Ç—Ä –∫–ª–∞—Å—Ç–µ—Ä–∞ –ø–µ—Ä–µ–º–µ—â–∞–µ—Ç—Å—è –≤ —Å—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ –≤–µ–∫—Ç–æ—Ä–æ–≤ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –æ–±—ä–µ–∫—Ç–æ–≤. –í –≤–∞—Ä–∏–∞–Ω—Ç–µ –ú–∞–∫ –ö–∏–Ω–∞ K-Means –Ω–∞—Ö–æ–¥–∏—Ç –ª–æ–∫–∞–ª—å–Ω—ã–π –º–∏–Ω–∏–º—É–º $\Phi_0$

–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, —Ü–µ–ª—å –∞–ª–≥–æ—Ä–∏—Ç–º–∞ - –≤—ã–±—Ä–∞—Ç—å —Ü–µ–Ω—Ç—Ä–æ–∏–¥—ã, –∫–æ—Ç–æ—Ä—ã–µ –º–∏–Ω–∏–º–∏–∑–∏—Ä—É—é—Ç *–∏–Ω–µ—Ä—Ü–∏—é*, –∏–ª–∏ –≤–Ω—É—Ç—Ä–∏–∫–ª–∞—Å—Å—Ç–µ—Ä–Ω—É—é –¥–∏—Å–ø–µ—Ä—Å—Å–∏—é (–ø–æ–º–Ω–∏—Ç–µ —Ñ–æ—Ä–º—É–ª—É –¥–∏—Å–ø–µ—Ä—Å–∏–∏? üòè):

$$J(C) = \sum_{i=0}^{n} \min_{\mu_j \in C}(||x_i - \mu_j||^2)$$

$C$ - –º–Ω–æ–∂–µ—Å—Ç–≤–æ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤, $\mu_j$ - —Ü–µ–Ω—Ç—Ä–∞ –∫–ª–∞—Å—Ç–µ—Ä–∞ $C_j$

**_–ù–∞ –ø—Ä–∏–º–µ—Ä–µ –æ–¥–Ω–æ–º–µ—Ä–Ω–æ–≥–æ —Å–ª—É—á–∞—è_**

–•–æ—Ç–∏–º –Ω–∞–π—Ç–∏:

$\underset{\mu}{\operatorname{argmin}} \frac{1}{N} \sum_{i=1}^{N}(x_i - \mu)^2$

$\frac{d}{d\mu}\frac{1}{N} \sum_{i=1}^{N}(x_i - \mu)^2 = \frac{2}{N} \sum_{i=1}^{N}(x_i - \mu) = 0 => \mu = \frac{1}{N} \sum_{i=1}^{N}x_i$

–ò–º–µ–Ω–Ω–æ –ø–æ—ç—Ç–æ–º—É –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ –Ω–µ –ï–≤–∫–ª–∏–¥–µ–≤–æ–π –º–µ—Ç—Ä–∏–∫–∏ –Ω–µ —Å–æ–≤—Å–µ–º –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ. –ù–∞–ø—Ä–∏–º–µ—Ä, –µ—Å–ª–∏ –µ—Å—Ç—å –∂–µ–ª–∞–Ω–∏–µ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –º–∞–Ω—Ö—ç—Ç—Ç–µ–Ω—Å–∫–æ–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ, —Ç–æ —ç—Ç–æ —É–∂–µ –Ω–µ k-means, –∞ k-medians

**–ó–∞–º–µ—á–∞–Ω–∏–µ:**

–ü—Ä–æ–±–ª–µ–º–∞ *–∏–Ω–µ—Ä—Ü–∏–∏* –≤ —Ç–æ–º –≤–∏–¥–µ, –≤ –∫–æ—Ç–æ—Ä–æ–º –æ–Ω–∞ –µ—Å—Ç—å –∑–∞–∫–ª—é—á–∞—á–µ—Ç—Å—è –≤ —Ç–æ–º, —á—Ç–æ –≤—ã–≥–æ–¥–Ω–µ–µ –≤—Å–µ–≥–æ –æ—Ç–Ω–µ—Å—Ç–∏ –∫–∞–∂–¥—ã–π –æ–±—ä–µ–∫—Ç –∫ —Å–≤–æ–µ–º—É –∫–ª–∞—Å—Ç–µ—Ä—É. –î–ª—è –ø—Ä–µ–¥–æ—Ç–≤—Ä–∞—â–µ–Ω–∏—è —Ç–∞–∫–æ–≥–æ, –Ω–∞–º —Å—Ç–æ–∏—Ç –≤—ã–±—Ä–∞—Ç—å —á–∏—Å–ª–æ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤, –ø–æ—Å–ª–µ –∫–æ—Ç–æ—Ä–æ–≥–æ –∏–Ω–µ—Ä—Ü–∏—è —É–±—ã–≤–∞–µ—Ç –Ω–µ —Å–ª–∏—à–∫–æ–º –±—ã—Å—Ç—Ä–æ. –ë–æ–ª–µ–µ —Ñ–æ—Ä–º–∞–ª—å–Ω–æ:

$$D(k) = \frac{|J(C_k) - J(C_{k+1})|}{|J(C_{k-1}) - J(C_{k})|} \rightarrow \min_{k}$$

```python
#create random centroids
np.random.seed(seed=SEED)
centroids = np.random.normal(loc=-2, scale=2, size=num_clusters*2)
centroids = centroids.reshape((3, 2))

cent_history = []
cent_history.append(centroids)

for i in range(10):
    #distanse between objects and centroids
    distances = cdist(X, centroids)
    labels = distances.argmin(axis=1)
    
    #rerange centroids coordinates
    centroids = centroids.copy()
    centroids[0, :] = np.mean(X[labels == 0, :], axis=0)
    centroids[1, :] = np.mean(X[labels == 1, :], axis=0)
    centroids[2, :] = np.mean(X[labels == 2, :], axis=0)

    cent_history.append(centroids)
    
plt.figure(figsize=(16, 16))
for i in range(4):
    distances = cdist(X, cent_history[i])
    labels = distances.argmin(axis=1)

    plt.subplot(2, 2, i + 1)
    plt.scatter(X[labels == 0, 0], X[labels == 0, 1], c='g', label='cluster 1')
    plt.scatter(X[labels == 1, 0], X[labels == 1, 1], c='c', label='cluster 2')
    plt.scatter(X[labels == 2, 0], X[labels == 2, 1], c='b', label='cluster 3')
    plt.plot(cent_history[i][:, 0], cent_history[i][:, 1], 'rX')
    plt.legend(loc='best')
    plt.title(f'Step {i+1}');

```

 P.S. –ó–∞—á–∞—Å—Ç—É—é, –∫–æ–≥–¥–∞ —Ä–µ—á—å –∏–¥–µ—Ç –æ–± –∏–µ—Ä–∞—Ä—Ö–∏—á–µ—Å–∫–æ–π –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏, —Ç–æ –ø–æ–¥—Ä–∞–∑—É–º–µ–≤–∞–µ—Ç—Å—è –∏–º–µ–Ω–Ω–æ –∞–≥–ª–æ–º–µ—Ä–∞—Ç–∏–≤–Ω–∞—è

**–í–æ–ø—Ä–æ—Å:** –∫–∞–∫ –≤—ã—á–∏—Å–ª—è—Ç—å —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è –º–µ–∂–¥—É –∫–ª–∞—Å—Ç–µ—Ä–∞–º–∏?

**–§–æ—Ä–º—É–ª–∞ –õ–∞–Ω—Å–∞-–£–∏–ª—å—è–º—Å–∞:**
$$R(U \cup V, S) = \alpha_U R(U, S) + \alpha_v R(V, S) + \beta R(U, V) + \gamma |R(U, S) - R(V, S)|$$

-—Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –º–µ–∂–¥—É –∫–ª–∞—Å—Ç–µ—Ä–æ–º, –æ–±—Ä–∞–∑–æ–≤–∞–Ω–Ω—ã–º –¥–≤—É–º—è —Å–ª–∏—Ç—ã–º–∏ $U, V$ –≤ –æ–¥–∏–Ω –∏ –∫–∞–∫–∏–º-—Ç–æ —Ç—Ä–µ—Ç—å–∏–º $S$. –í–∞—Ä—å–∏—Ä—É—è –ø–∞—Ä–∞–º–µ—Ç—Ä—ã $\alpha_U, \alpha_V, \beta, \gamma$ –º–æ–∂–Ω–æ –ø–æ–ª—É—á–∏—Ç—å —Ä–∞–∑–ª–∏—á–Ω—ã–µ —Å–ø–æ—Å–æ–±—ã –≤—ã—á–∏—Å–ª–µ–Ω–∏—è —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–π –º–µ–∂–¥—É –∫–ª–∞—Å—Ç–µ—Ä–∞–º–∏.

–ù–∞–ø—Ä–∏–º–µ—Ä,

Single linkage - –º–∏–Ω–∏–º—É–º –ø–æ–ø–∞—Ä–Ω—ã—Ö —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–π –º–µ–∂–¥—É —Ç–æ—á–∫–∞–º–∏ –∏–∑ –¥–≤—É—Ö –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ ($\alpha_U = \alpha_V = \frac{1}{2}, \beta = 0, \gamma = -\frac{1}{2}$)

$$d(W, S) = \min_{w \in W, s \in S} \rho(w, s)$$
  
Complete linkage ‚Äî –º–∞–∫—Å–∏–º—É–º –ø–æ–ø–∞—Ä–Ω—ã—Ö —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–π –º–µ–∂–¥—É —Ç–æ—á–∫–∞–º–∏ –∏–∑ –¥–≤—É—Ö –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ ($\alpha_U = \alpha_V = \frac{1}{2}, \beta = 0, \gamma = \frac{1}{2}$)

$$d(W, S) = \max_{w \in W, s \in S} \rho(w, s)$$

Average linkage ‚Äî —Å—Ä–µ–¥–Ω–µ–µ –ø–æ–ø–∞—Ä–Ω—ã—Ö —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–π –º–µ–∂–¥—É —Ç–æ—á–∫–∞–º–∏ –∏–∑ –¥–≤—É—Ö –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ ($\alpha_U = \frac{|U|}{|W|}, \alpha_V = \frac{|V|}{|W|}, \beta = \gamma = 0$)

$$d(W, S) = \frac{1}{|W||S|} \sum_{w \in W} \sum_{s \in S}\rho(w, s)$$

Centroid linkage ‚Äî —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –º–µ–∂–¥—É —Ü–µ–Ω—Ç—Ä–∞–º–∏ –¥–≤—É—Ö –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ ($\alpha_U = \frac{|U|}{|W|}, \alpha_V = \frac{|V|}{|W|}, \beta = -\alpha_U\alpha_V, \gamma = 0$)

$$d(W, S) = \rho(\sum_{w \in W} \frac{w}{|W|}, \sum_{s \in S} \frac{s}{|S|})$$

### –î–µ–Ω–¥–æ–≥—Ä–∞–º–º–∞

–î–µ–Ω–¥–æ–≥—Ä–∞–º–º–∞, –∏–ª–∏ –¥–µ—Ä–µ–≤–æ —Å–∫–ª–µ–∏–≤–∞–Ω–∏—è, –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç –∏—Ç–µ—Ä–∞—Ç–∏–≤–Ω—ã–π –ø—Ä–æ—Ü–µ—Å—Å –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏—è –∫–ª–∞—Å—Ç–µ—Ä–æ–≤, –ø—Ä–∏—á–µ–º —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –º–µ–∂–¥—É –∫–ª–∞—Å—Ç–µ—Ä–∞–º–∏ –æ—Ç–æ–±—Ä–∞–∂–∞–µ—Ç—Å—è –∫–∞–∫ –≤—ã—Å–æ—Ç–∞ –¥—É–≥–∏, –∫–æ—Ç–æ—Ä–æ–π —Å–æ–µ–¥–∏–Ω—è—é—Ç—Å—è —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–µ –º–µ—Ç–∫–∏ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤. 

–ë–ª–∞–≥–æ–¥–∞—Ä—è –¥–µ–Ω–¥–æ–≥—Ä–∞–º–º–µ –µ—Å—Ç—å —Ç–∞–∫–∂–µ –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç—å –ø–æ–¥–±–∏—Ä–∞—Ç—å –æ–ø—Ç–∏–º–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤

```python
def plot_dendrogram(model, **kwargs):
    '''Create linkage matrix and then plot the dendrogram
    create the counts of samples under each node
    '''
    
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack([model.children_, model.distances_, counts]).astype(float)

    # Plot the corresponding dendrogram
    dendrogram(linkage_matrix, **kwargs)
```

```python
agclust = AgglomerativeClustering(n_clusters=None, distance_threshold=1).fit(X)
```

```python
plt.figure(figsize=(7, 7))
plot_dendrogram(agclust)
```

–¥–æ –¥—Ä—É–≥–æ–π –±–ª–∏–∂–∞–π—à–µ–π —Ç–æ—á–∫–∏ –∏–∑ –≤—ã–±–æ—Ä–∫–∏. –ï—Å–ª–∏ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –ø–æ–ª—É—á–∞–µ—Ç—Å—è –±–ª–∏–∑–∫–æ–π –∫ 1/2, —ç—Ç–æ –∑–Ω–∞—á–∏—Ç, —á—Ç–æ –≤—ã–±–æ—Ä–∫–∞ –±–æ–ª–µ–µ –∏–ª–∏ –º–µ–Ω–µ–µ —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ –∑–∞–ø–æ–ª–Ω—è–µ—Ç –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤. –ï—Å–ª–∏ –∂–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –ø–æ–ª—É—á–∞–µ—Ç—Å—è –≤–±–ª–∏–∑–∏ –Ω—É–ª—è, —ç—Ç–æ –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ —Ç–æ—á–∫–∏ –∫–∞–∫-—Ç–æ –≥—Ä—É–ø–ø–∏—Ä—É—é—Ç—Å—è.

$$H = \frac{\sum_{i=1}^{n} w_i}{\sum_{i=1}^{n} w_i + \sum_{i=1}^{n} u_i}$$



```python
def hopkins_statistics(X):
    '''Calculate hopkins statistics to check cluster stucture in data
    
    Parameters:
        X - numpy matrix
        
    Return:
        statistics
        
    Note:
        Estimation the clusterability of a dataset. A score between 0 and 1, a score around 0.5 express
        no clusterability and a score tending to 0 express a high cluster tendency
    '''
    n, d = X.shape[0], X.shape[1]
    m = int(0.1 * n) # heuristic from article
    
    nbrs = NearestNeighbors(n_neighbors=1).fit(X)
 
    rand_X = sample(range(n), m)
 
    ujd = []
    wjd = []
    for j in range(m):
        u_dist, _ = nbrs.kneighbors(uniform(np.amin(X,axis=0), np.amax(X,axis=0), d).reshape(1, -1)
                                    , 2, return_distance=True)
        ujd.append(u_dist[0][1])
        
        w_dist, _ = nbrs.kneighbors(X[rand_X[j]].reshape(1, -1), 2, return_distance=True)
        wjd.append(w_dist[0][1])
 
    H = sum(wjd) / (sum(ujd) + sum(wjd))
    
    return H
```

```python
print(f'Hopkins statistics: {hopkins_statistics(X)}')
```

—Ç–∞–∫ –∂–µ –º—ã —Ö–æ—Ç–∏–º, —á—Ç–æ–±—ã –æ–±—ä–µ–∫—Ç—ã –∏–∑ —Ä–∞–∑–Ω—ã—Ö –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ –±—ã–ª–∏ –∫–∞–∫ –º–æ–∂–Ω–æ –¥–∞–ª—å—à–µ –¥—Ä—É–≥ –æ—Ç –¥—Ä—É–≥–∞, **–º–µ–∂–∫–ª–∞—Å—Ç–µ—Ä–Ω–æ–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ**: $F_1 = \frac{\sum_{i<j}[y_i\neq y_j]\rho(x_i; x_j)}{\sum_{i<j}[y_i\neq y_j]} \rightarrow max$.

–ù–∞—á–Ω–µ–º —Å –º–µ—Ç–æ–¥–∞ –≥–ª–∞–≤–Ω—ã—Ö –∫–æ–º–ø–æ–Ω–µ–Ω—Ç.

### PCA

PCA (principal component analysis, –º–µ—Ç–æ–¥ –≥–ª–∞–≤–Ω—ã—Ö –∫–æ–º–ø–æ–Ω–µ–Ω—Ç). –ï—Å—Ç—å –Ω–µ—Å–∫–æ–ª—å–∫–æ –ø–æ–¥—Ö–æ–¥–æ–≤ –∫ –ø–æ—Å—Ç–∞–Ω–æ–≤–∫–µ –∑–∞–¥–∞—á–∏ –º–µ—Ç–æ–¥–∞ –≥–ª–∞–≤–Ω—ã—Ö –∫–æ–º–ø–æ–Ω–µ–Ω—Ç, —Ä–∞—Å—Å–º–æ—Ç—Ä–∏–º –¥–≤–∞ –∏–∑ –Ω–∏—Ö: 
1. –∫–∞–∫ –ª–∏–Ω–µ–π–Ω—ã–π –º–µ—Ç–æ–¥ –ø–æ–Ω–∏–∂–µ–Ω–∏—è —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–∏.

**–ò—Å—Ö–æ–¥–Ω–∞—è –∑–∞–¥–∞—á–∞:** –Ω–∞–π—Ç–∏ —Ç–∞–∫—É—é –º–∞—Ç—Ä–∏—Ü—É $Z$, —á—Ç–æ–±—ã $Z=XW^T$
<br>
–î–ª—è —Å—É—â–µ—Å—Ç–≤–æ–≤–∞–Ω–∏—è –µ–¥–∏–Ω—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ —Ä–µ—à–µ–Ω–∏—è —Ç—Ä–µ–±—É–µ—Ç—Å—è –≤–≤–µ—Å—Ç–∏ –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–µ $W^{T}W=E$, —Ç–æ–≥–¥–∞ $X=ZW$. –ò —Ä–µ—à–∞–µ—Ç—Å—è –∑–∞–¥–∞—á–∞ –º–∞—Ç—Ä–∏—á–Ω–æ–≥–æ —Ä–∞–∑–ª–æ–∂–µ–Ω–∏—è –ø–æ –Ω–æ—Ä–º–µ –§—Ä–æ–±–µ–Ω–∏—É—Å–∞ $||X-ZW||^2 \rightarrow \min_{Z, W}$. –ù–µ–æ–±—Ö–æ–¥–∏–º–æ –ø—Ä–µ–¥—Å—Ç–∞–≤–∏—Ç—å –º–∞—Ç—Ä–∏—Ü—É $X$ –≤ –≤–∏–¥–µ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏—è –¥–≤—É—Ö –º–∞—Ç—Ä–∏—Ü $Z, W$ –º–µ–Ω—å—à–µ–≥–æ —Ä–∞–Ω–≥–∞.
2. –∫–∞–∫ –º–∞–∫—Å–∏–º–∏–∑–∞—Ü–∏—è –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –≤—ã–±–æ—Ä–∫–∏ –ø–æ—Å–ª–µ –ø–æ–Ω–∏–∂–µ–Ω–∏—è —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–∏. –ü—É—Å—Ç—å –µ—Å—Ç—å –≤—ã–±–æ—Ä–∫–∞ –∏ —Ç—Ä–µ–±—É–µ—Ç—Å—è –≤—ã–±—Ä–∞—Ç—å –ø—Ä—è–º—ã–µ (–ø—Ä—è–º—É—é), –Ω–∞–∑—ã–≤–∞–µ–º—ã–µ –≥–ª–∞–≤–Ω—ã–º–∏ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç–∞–º–∏, –Ω–∞ –∫–æ—Ç–æ—Ä—ã–µ —Å–ø—Ä–æ–µ—Ü–∏—Ä–æ–≤–∞—Ç—å –∏—Å—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –æ–ø—Ç–∏–º–∞–ª—å–Ω–µ–π –≤—Å–µ–≥–æ —Å —Ç–æ—á–∫–∏ –∑—Ä–µ–Ω–∏—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è –∫–∞–∫ –º–æ–∂–Ω–æ –±–æ–ª—å—à–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏. –í –¥–∞–Ω–Ω–æ–º —Å–ª—É—á–∞–µ –ø–æ–¥ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–µ–π –∏–º–µ–µ—Ç—Å—è –≤ –≤–∏–¥—É –¥–∏—Å–ø–µ—Ä—Å–∏—è, —Ç–æ –µ—Å—Ç—å —á–µ–º –±–æ–ª—å—à–µ –¥–∏—Å–ø–µ—Ä—Å–∏—è –≤—ã–±–æ—Ä–∫–∏ –ø–æ—Å–ª–µ –ø—Ä–æ–µ—Ü–∏—Ä–æ–≤–∞–Ω–∏—è, —Ç–µ–º –ª—É—á—à–µ - —Ç–µ–º –±–æ–ª—å—à–µ —Å–æ—Ö—Ä–∞–Ω—è–µ—Ç—Å—è –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏.

    –§–æ—Ä–º–∞–ª—å–Ω–æ –¥–∏—Å–ø–µ—Ä—Å–∏—è –≤—ã–±–æ—Ä–∫–∏ –ø–æ—Å–ª–µ –ø—Ä–æ–µ—Ü–∏—Ä–æ–≤–∞–Ω–∏—è –∑–∞–ø–∏—Å—ã–≤–∞–µ—Ç—Å—è —Ç–∞–∫: $\sum_{i=1}^{d} w_{i}^TX^TXw_{i} \rightarrow max_{W}$. –ß–µ–º –±–æ–ª—å—à–µ –∑–Ω–∞—á–µ–Ω–∏–µ —ç—Ç–æ–π —Å—É–º–º—ã, —Ç–µ–º –±–æ–ª—å—à–µ –æ–∫–∞–∑—ã–≤–∞–µ—Ç—Å—è –¥–∏—Å–ø–µ—Ä—Å–∏—è –≤—ã–±–æ—Ä–∫–∏ –ø–æ—Å–ª–µ –ø—Ä–æ–µ—Ü–∏—Ä–æ–≤–∞–Ω–∏—è –Ω–∞ –≥–∏–ø–µ—Ä–ø–ª–æ—Å–∫–æ—Å—Ç—å, –∫–æ—Ç–æ—Ä–∞—è –∑–∞–¥–∞–µ—Ç—Å—è –º–∞—Ç—Ä–∏—Ü–µ–π –≤–µ—Å–æ–≤ $W$.
    
    **–í –∏—Ç–æ–≥–µ –∏–º–µ–µ–º –∑–∞–¥–∞—á—É:**
    <br>
    $$\sum_{i=1}^{d} w_{i}^TX^TXw_{i} \rightarrow \max_{W},$$ –ø—Ä–∏ —É—Å–ª–æ–≤–∏–∏
    $$W^{T}W=E$$
    
   **–†–µ—à–µ–Ω–∏–µ:** —á–µ—Ä–µ–∑ –º–µ—Ç–æ–¥ –º–Ω–æ–∂–∏—Ç–µ–ª–µ–π –õ–∞–≥—Ä–∞–Ω–∂–∞

**–ó–∞–º–µ—á–∞–Ω–∏–µ**: –≤ –º–µ—Ç–æ–¥–µ –≥–ª–∞–≤–Ω—ã—Ö –∫–æ–º–ø–æ–Ω–µ–Ω—Ç –≤—ã—Ä–∞–∂–µ–Ω–∏–µ, –æ–ø–∏—Å—ã–≤–∞—é—â–µ–µ –¥–∏—Å–ø–µ—Ä—Å–∏—é –æ–∑–Ω–∞—á–∞–µ—Ç –∏–º–µ–Ω–Ω–æ –¥–∏—Å–ø–µ—Ä—Å–∏—é –≤—ã–±–æ—Ä–∫–∏, –µ—Å–ª–∏ –º–∞—Ç—Ä–∏—Ü–∞ –æ–±—ä–µ–∫—Ç—ã-–ø—Ä–∏–∑–Ω–∞–∫–∏ **—Ü–µ–Ω—Ç—Ä–∏—Ä–æ–≤–∞–Ω–∞**

–†–µ—à–∞—è –¥–∞–Ω–Ω—É—é –∑–∞–¥–∞—á—É, –º–æ–∂–Ω–æ —É–≤–∏–¥–µ—Ç—å, —á—Ç–æ –ø–æ—Å–∫–æ–ª—å–∫—É —Ç—Ä–µ–±—É–µ—Ç—Å—è –º–∞–∫—Å–∏–º–∏–∑–∏—Ä–æ–≤–∞—Ç—å –¥–∏—Å–ø–µ—Ä—Å–∏—é, –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –≤—ã–±–∏—Ä–∞—Ç—å –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∏ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–π –≤–µ–∫—Ç–æ—Ä, –∫–æ—Ç–æ—Ä—ã–π —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç —ç—Ç–æ–º—É –∑–Ω–∞—á–µ–Ω–∏—é.
–ò—Ç–∞–∫, –≤ –º–µ—Ç–æ–¥–µ –≥–ª–∞–≤–Ω—ã—Ö –∫–æ–º–ø–æ–Ω–µ–Ω—Ç –ø–µ—Ä–≤–∞—è –∫–æ–º–ø–æ–Ω–µ–Ω—Ç–∞ ‚Äî —ç—Ç–æ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–π –≤–µ–∫—Ç–æ—Ä –º–∞—Ç—Ä–∏—Ü—ã $X^TX$, –∫–æ—Ç–æ—Ä—ã–π —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–º—É —Å–æ–±—Å—Ç–≤–µ–Ω–Ω–æ–º—É –∑–Ω–∞—á–µ–Ω–∏—é —ç—Ç–æ–π –º–∞—Ç—Ä–∏—Ü—ã. $X^TX$ - –º–∞—Ç—Ä–∏—Ü–∞ –∫–æ–≤–∞—Ä–∏–∞—Ü–∏–∏

–ù–∞–π–¥—è $d$ –≥–ª–∞–≤–Ω—ã—Ö –∫–æ–º–ø–æ–Ω–µ–Ω—Ç $w_1, w_2, ..., w_d$ - —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ –≤–µ–∫—Ç–æ—Ä—ã $X^{T}X$, $\lambda_1, \lambda_2, ..., \lambda_d$ - –∏—Ö —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è $=>$ –º–æ–∂–µ–º —É–∑–Ω–∞—Ç—å –¥–æ–ª—é —Å–æ—Ö—Ä–∞–Ω–∏–≤—à–µ–π—Å—è –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –≤ –¥–∞–Ω–Ω—ã—Ö: $\frac{\sum_{i=1}^{d} \lambda_i}{\sum_{i=1}^{D}\lambda_i}$

*PCA —Å—Ç—Ä–µ–º–∏—Ç—Å—è —Å–æ—Ö—Ä–∞–Ω–∏—Ç—å –≥–ª–æ–±–∞–ª—å–Ω—É—é —Å—Ç—Ä—É–∫—Ç—É—Ä—É –≤ –¥–∞–Ω–Ω—ã—Ö*

**–ö—Ä–∞—Ç–∫–∏–π –∏—Ç–æ–≥ —Ä–∞–±–æ—Ç—ã PCA**

–°–æ–≤–µ—Ä—à–∞–µ–º –ø–µ—Ä–µ—Ö–æ–¥ –∫ –Ω–æ–≤—ã–º –æ—Å—è–º, —Ç–∞–∫ —á—Ç–æ:
    
    –ù–æ–≤—ã–µ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ —è–≤–ª—è—é—Ç—Å—è –ª–∏–Ω–µ–π–Ω–æ–π –∫–æ–º–±–∏–Ω–∞—Ü–∏–µ–π —Å—Ç–∞—Ä—ã—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö
    –ù–æ–≤—ã–µ –æ—Å–∏ –æ—Ä—Ç–æ–≥–æ–Ω–∞–ª—å–Ω—ã –¥—Ä—É–≥ –¥—Ä—É–≥—É
    –î–∏—Å–ø–µ—Ä—Å–∏—è –≤–¥–æ–ª—å –Ω–æ–≤—ã—Ö –æ—Å–µ–π –º–∞–∫—Å–∏–º–∞–ª—å–Ω–∞—è

**–ü–æ—Å–º–æ—Ç—Ä–∏–º, –∫–∞–∫ —Å—Ç—Ä–æ—è—Ç—Å—è –≥–ª–∞–≤–Ω—ã–µ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç—ã**

```python
import matplotlib.patches as mpatches
from sklearn.decomposition import PCA
```

```python
def plot_principal_components(data, model, scatter=True, legend=True):
    '''plot princial components'''
    
    W_pca = model.components_
    if scatter:
        plt.scatter(data[:,0], data[:,1])
    plt.plot(data[:,0], -(W_pca[0,0]/W_pca[0,1])*data[:,0], color="c")
    plt.plot(data[:,0], -(W_pca[1,0]/W_pca[1,1])*data[:,0], color="c")
    if legend:
        c_patch = mpatches.Patch(color='c', label='Principal components')
        plt.legend(handles=[c_patch], loc='lower right')
    # —Å–¥–µ–ª–∞–µ–º –≥—Ä–∞—Ñ–∏–∫–∏ –∫—Ä–∞—Å–∏–≤—ã–º–∏:
    plt.axis('equal')
    limits = [np.minimum(np.amin(data[:,0]), np.amin(data[:,1]))-0.5,
              np.maximum(np.amax(data[:,0]), np.amax(data[:,1]))+0.5]
    plt.xlim(limits[0],limits[1])
    plt.ylim(limits[0],limits[1])
    plt.draw()
```

```python
mu = np.ones(2)
sigma = np.array([[3,1],[1,2]]) #–º–∞—Ç—Ä–∏—Ü–∞ –∫–æ–≤–∞—Ä–∏–∞—Ü–∏–∏

data = np.random.multivariate_normal(mu, sigma, size=200)
plt.scatter(data[:,0], data[:,1])
plt.show()
```

```python
#by PCA
model = PCA(n_components=2, random_state=SEED)
model.fit(data)

#by hand
eigenvalue, W_orig = np.linalg.eig(sigma)

#plot
plt.figure(figsize=(7, 6))
plt.scatter(data[:,0], data[:,1])

#true components with max variance
plt.plot(data[:,0], (W_orig[0,0]/W_orig[0,1])*data[:,0], color="g")
plt.plot(data[:,0], (W_orig[1,0]/W_orig[1,1])*data[:,0], color="g")

#pca components
plot_principal_components(data, model, scatter=False, legend=False)
c_patch = mpatches.Patch(color='c', label='Principal components')
g_patch = mpatches.Patch(color='g', label='True components')

plt.legend(handles=[g_patch, c_patch])

plt.show()
```

## SVD

–°–∏–Ω–≥—É–ª—è—Ä–Ω–æ–µ —Ä–∞–∑–ª–æ–∂–µ–Ω–∏–µ –º–∞—Ç—Ä–∏—Ü—ã $X$ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è–µ—Ç —Å–æ–±–æ–π:

$X = U \Lambda V^T$, –≥–¥–µ 
<br>
$U$, $V$ -  –æ—Ä—Ç–æ–≥–æ–Ω–∞–ª—å–Ω—ã–µ –º–∞—Ç—Ä–∏—Ü—ã, $\Lambda$ - –¥–∏–∞–≥–æ–Ω–∞–ª—å–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞;
<br>
–°–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ –≤–µ–∫—Ç–æ—Ä–∞ –º–∞—Ç—Ä–∏—Ü $XX^T$ –∏ $X^TX$ —Å—É—Ç—å —Å—Ç–æ–ª–±—Ü—ã –º–∞—Ç—Ä–∏—Ü $U$ –∏ $V$ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ, —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ –∂–µ –∑–Ω–∞—á–µ–Ω–∏—è —ç—Ç–∏—Ö –º–∞—Ç—Ä–∏—Ü, –≤–æ–æ–±—â–µ –≥–æ–≤–æ—Ä—è, —Å–æ–≤–ø–∞–¥–∞—é—Ç –∏ —Å—Ç–æ—è—Ç –Ω–∞ –¥–∏–∞–≥–æ–Ω–∞–ª–∏ –º–∞—Ç—Ä–∏—Ü—ã $\Lambda$ –∏ –Ω–∞–∑—ã–≤–∞—é—Ç—Å—è **—Å–∏–Ω–≥—É–ª—è—Ä–Ω—ã–º–∏ —á–∏—Å–ª–∞–º–∏**.
<br>
–ê –ø—Ä–∏—á–µ–º —Ç—É—Ç PCA?
1. –ù–∞—Ö–æ–¥–∏–º —Å–∏–Ω–≥—É–ª—è—Ä–Ω–æ–µ —Ä–∞–∑–ª–æ–∂–µ–Ω–∏–µ –º–∞—Ç—Ä–∏—Ü—ã $X$
2. –§–æ—Ä–º—É—Ä—É–µ–º –º–∞—Ç—Ä–∏—Ü—É –≤–µ—Å–æ–≤ $W$ –∏–∑ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã—Ö –≤–µ–∫—Ç–æ—Ä–æ–≤ (—Å—Ç–æ–ª–±—Ü—ã –º–∞—Ç—Ä–∏—Ü—ã $V$, —Ç–∞–∫ –∫–∞–∫ –¥–ª—è –º–∞—Ç—Ä–∏—Ü—ã $X^TX$, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏—Ö –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–º —Å–∏–Ω–≥—É–ª—è—Ä–Ω—ã–º —á–∏—Å–ª–∞–º)
3. –ü–µ—Ä–µ—Ö–æ–¥–∏–º –≤ –Ω–æ–≤–æ–µ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ: $Z=XW$

–ù–∞ –ø—Ä–∞–∫—Ç–∏–∫–µ –∑–∞—á–∞—Å—Ç—É—é –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –∏–º–µ–Ω–Ω–æ —Å–∏–Ω–≥—É–ª—è—Ä–Ω–æ–µ —Ä–∞–∑–ª–æ–∂–µ–Ω–∏–µ, —Ç–∞–∫ –∫–∞–∫ –æ–Ω–æ –≤—ã—á–∏—Å–ª–∏—Ç–µ–ª—å–Ω–æ –±—ã—Å—Ç—Ä–µ–µ –∏ –º–µ—Ç–æ–¥ –≥–ª–∞–≤–Ω—ã—Ö –∫–æ–º–ø–æ–Ω–µ–Ω—Ç —Ä–µ–∞–ª–∏–∑—É–µ—Ç—Å—è —á–µ—Ä–µ–∑ –Ω–µ–≥–æ. –í Sklearn —Ç–∞–∫–∂–µ –µ—Å—Ç—å truncatedSVD, –ø–æ–∑–≤–æ–ª—è—é—â–∏–π —Ä–∞–±–æ—Ç–∞—Ç—å —Å —Ä–∞–∑—Ä–µ–∂–µ–Ω–Ω—ã–º–∏ –º–∞—Ç—Ä–∏—Ü–∞–º–∏, —á—Ç–æ –≤–µ—Å—å–º–∞ –≤–∞–∂–Ω–æ, –Ω–∞–ø—Ä–∏–º–µ—Ä, –ø—Ä–∏ —Ä–∞–∑–ª–æ–∂–µ–Ω–∏–∏ –º–∞—Ç—Ä–∏—Ü—ã [–∫–ª–∏–µ–Ω—Ç—ã-–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–∫—É–ø–æ–∫] –≤ –ø—Ä–æ–¥—É–∫—Ç–æ–≤–æ–º –º–∞–≥–∞–∑–∏–Ω–µ, —Ç–∞–∫ –∫–∞–∫ —Ç–æ–≤–∞—Ä–æ–≤ –º–Ω–æ–≥–æ, –Ω–æ –¥–∞–ª–µ–∫–æ –Ω–µ –∫–∞–∂–¥—ã–π –ø–æ–∫—É–ø–∞–µ—Ç—Å—è –æ—á–µ—Ä–µ–¥–Ω—ã–º –∫–ª–∏–µ–Ω—Ç–æ–º. –¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º –º–æ–∂–Ω–æ –¥–µ–ª–∞—Ç—å —Ä–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–∏ –¥–ª—è –∫–ª–∏–µ–Ω—Ç–æ–≤, –Ω–æ –æ–± —ç—Ç–æ–º –Ω–µ —Å–µ–≥–æ–¥–Ω—è üò≤

## t-SNE

- $||x_i - x_j||$ - –µ–≤–∫–ª–∏–¥–æ–≤–∞ –º–µ—Ç—Ä–∏–∫–∞
- ${x_i}$ - –æ–±—ä–µ–∫—Ç—ã –≤ –∏—Å—Ö–æ–¥–Ω–æ–º –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ
- $\widetilde{x_i}$ - –æ–±—ä–µ–∫—Ç—ã –≤ –º–∞–ª–æ–º–µ—Ä–Ω–æ–º –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ

–¢—Ä–µ–±—É–µ—Ç—Å—è —Å–æ—Ö—Ä–∞–Ω–∏—Ç—å –ø—Ä–æ–ø–æ—Ä—Ü–∏–∏ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–π –≤ –º–∞–ª–æ–º–µ—Ä–Ω–æ–º –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å–≤–µ –ø—Ä–∏ –ø–µ—Ä–µ—Ö–æ–¥–µ –∏–∑ –º–Ω–æ–≥–æ–º–µ—Ä–Ω–æ–≥–æ (–ø–æ—Ç–æ–º—É —á—Ç–æ —Å–æ—Ö—Ä–∞–Ω–∏—Ç—å –ø—Ä–∏ –±–æ–ª—å—à–æ–º —É–º–µ–Ω—å—à–µ–Ω–∏–∏ —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–∏ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–∞ –ø–æ–ø–∞—Ä–Ω—ã–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –∫—Ä–∞–π–Ω–µ —Å–ª–æ–∂–Ω–æ - **–º–µ—Ç–æ–¥ –º–Ω–æ–≥–æ–º–µ—Ä–Ω–æ–≥–æ —à–∫–∞–ª–∏—Ä–æ–≤–∞–Ω–∏—è**), —Ç–æ –µ—Å—Ç—å

–µ—Å–ª–∏ $$\rho(x_i, x_j) = \alpha\rho(x_i, x_k),$$ —Ç–æ 
$$\rho(\widetilde{x_i}, \widetilde{x_j}) = \alpha\rho(\widetilde{x_i}, \widetilde{x_k})$$

–†–∞—Å—Å—Ç–æ—è–Ω–∏–µ –≤ –∏—Å—Ö–æ–¥–Ω–æ–º –∏ –º–∞–ª–æ–º–µ—Ä–Ω–æ–º –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–∞—Ö:

$$p(x_j|x_i) = \frac{\exp(||x_i - x_j||^2 / 2 \sigma^2)}{\sum_{i\neq k}\exp(||x_i - x_k||^2 / 2 \sigma^2)}$$

$$q(\widetilde{x_j}|\widetilde{x_i}) = \frac{(1+||\widetilde{x_i}-\widetilde{x_j}||^2)^{-1}}{\sum_{i\neq k}(1+||\widetilde{x_i}-\widetilde{x_k}||^2)^{-1}}$$

–ï—Å–ª–∏ –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ $q(\widetilde{x_j}|\widetilde{x_i})$ —Å–¥–µ–ª–∞—Ç—å —Ç–∞–∫–∏–º –∂–µ, –∫–∞–∫ –∏ $p(x_j|x_i)$, —Ç–æ –ø–æ–ª—É—á–∏–º –º–µ—Ç–æ–¥ SNE, –Ω–µ–¥–æ—Å—Ç–∞—Ç–∫–æ–º –∫–æ—Ç–æ—Ä–æ–≥–æ —è–≤–ª—è—é—Ç—Å—è –±–æ–ª–µ–µ —Å–∏–ª—å–Ω—ã–µ —à—Ç—Ä–∞—Ñ—ã –∑–∞ –Ω–µ—Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –ø—Ä–æ–ø–æ—Ä—Ü–∏–π (–≤–≤–∏–¥—É –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è –ï–≤–∫–ª–∏–¥–æ–≤–æ–π –º–µ—Ç—Ä–∏–∫–∏). –í–≤–∏–¥—É —Å—É—â–µ—Å—Ç–≤–æ–≤–∞–Ω–∏—è –ø—Ä–æ–±–ª–µ–º—ã –≤ –º–Ω–æ–≥–æ–º–µ—Ä–Ω—ã—Ö –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–∞—Ö, –Ω–∞–∑—ã–≤–∞–µ–º–æ–π **–ø—Ä–æ–∫–ª—è—Ç–∏–µ–º —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–∏** (–≤ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–∞—Ö –≤—ã—Å–æ–∫–æ–π —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–∏ –ª–µ–≥–∫–æ —Ä–∞—Å–ø–æ–ª–æ–∂–∏—Ç—å –æ–±—ä–µ–∫—Ç—ã –±–ª–∏–∑–∫–æ –¥—Ä—É–≥ –∫ –¥—Ä—É–≥—É, —Ç–∞–∫–∂–µ —á–µ–º –±–æ–ª—å—à–µ —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–∞, —Ç–µ–º –±–æ–ª–µ–µ –±–ª–∏–∑–∫–∏ –¥—Ä—É–≥ –∫ –¥—Ä—É–≥—É —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è –º–µ–∂–¥—É –ø–∞—Ä–∞–º–∏ –æ–±—ä–µ–∫—Ç–æ–º; –≤ –º–∞–ª–æ–º–µ—Ä–Ω—ã—Ö –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–∞—Ö —Å–æ—Ö—Ä–∞–Ω–∏—Ç—å —ç—Ç–æ —Å–≤–æ–π—Å—Ç–≤–æ —Å–ª–æ–∂–Ω–æ), –≥–æ—Ä–∞–∑–¥–æ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–µ–µ –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç —Å–µ–±—è t-SNE

–î–ª—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è –¥–≤—É—Ö —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–π –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –¥–∏–≤–µ—Ä–≥–µ–Ω—Ü–∏—è –ö—É–ª—å–±–∞–∫–∞-–õ–µ–π–±–ª–µ—Ä–∞, –∫–æ—Ç–æ—Ä—É—é –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –º–∏–Ω–∏–º–∏–∑–∏—Ä–æ–≤–∞—Ç—å (–Ω–∞–ø—Ä–∏–º–µ—Ä, –º–µ—Ç–æ–¥–æ–º —Å—Ç–æ—Ö–∞—Å—Ç–∏—á–µ—Å–∫–æ–≥–æ –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω–æ–≥–æ —Å–ø—É—Å–∫–∞):

$$D_{KL}(p||q) = \sum_{i=1}^{l}\sum_{i \neq j}p(x_j|x_i) \log\frac{p(x_j|x_i)}{q(\widetilde{x_j}|\widetilde{x_i})} \rightarrow \min_{\widetilde{x_1},...,\widetilde{x_l}}$$

–ù–∞–∏–±–æ–ª–µ–µ —á–∞—Å—Ç–æ–µ –ø—Ä–∏–º–µ–Ω–µ–Ω–∏–µ –º–µ—Ç–æ–¥ t-SNE –Ω–∞—Ö–æ–¥–∏—Ç –ø—Ä–∏ –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏ –¥–∞–Ω–Ω—ã—Ö –Ω–∞ –¥–≤—É–º–µ—Ä–Ω—É—é –∏–ª–∏ —Ç—Ä–µ—Ö–º–µ—Ä–Ω—É—é –≥–∏–ø–µ—Ä–ø–ª–æ—Å–∫–æ—Å—Ç–∏.

**–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ:** –µ—Å–ª–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å $q(\widetilde{x_j}|\widetilde{x_i})$ –≤—ã—á–∏—Å–ª—è—Ç—å —Ç–∞–∫–∂–µ –∫–∞–∫ –∏ $p(x_j|x_i)$, —Ç–æ –ø–æ–ª—É—á–∏–º –º–µ—Ç–æ–¥, –Ω–∞–∑—ã–≤–∞–µ–º—ã **SNE**, –Ω–æ –µ–≥–æ –±–µ–¥–∞ –∫–∞–∫ —Ä–∞–∑ –≤ —Ç–æ–º, —á—Ç–æ –≤ –º–∞–ª–æ–º–µ—Ä–Ω–æ–º –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ –æ–±—ä–µ–∫—Ç—ã —Å–ª–æ–∂–Ω–µ–µ —Ä–∞–∑–º–µ—Å—Ç–∏—Ç—å —Ä—è–¥–æ–º, —á–µ–º –≤ –º–Ω–æ–≥–æ–º–µ—Ä–Ω–æ–º (**–ø—Ä–æ–∫–ª—è—Ç–∏–µ —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–∏**), –∞ –ï–≤–∫–ª–∏–¥–æ–≤–∞ –º–µ—Ç—Ä–∏–∫–∞ —Å–ª–∏—à–∫–æ–º —Å–∏–ª—å–Ω–æ —à—Ç—Ä–∞—Ñ—É–µ—Ç –∑–∞ –Ω–µ—Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –ø—Ä–æ–ø–æ—Ä—Ü–∏–π. –ú–µ—Ç–æ–¥ **t-SNE** –∫–∞–∫ —Ä–∞–∑-—Ç–∞–∫–∏ —Å—Ç–∞—Ä–∞–µ—Ç—Å—è —Ä–µ—à–∏—Ç—å –¥–∞–Ω–Ω—É—é –ø—Ä–æ–±–ª–µ–º—É, –≤–≤–æ–¥—è $q(\widetilde{x_j}|\widetilde{x_i})$ –∏–Ω–∞—á–µ.

**–í–æ–ø—Ä–æ—Å:** –∑–∞—á–µ–º –º–æ–∂–µ—Ç –ø–æ–Ω–∞–¥–æ–±–∏—Ç—å—Å—è –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å –¥–∞–Ω–Ω—ã–µ?

### –ù–µ–º–Ω–æ–≥–æ –ø—Ä–∞–∫—Ç–∏–∫–∏ –ø—Ä–∏–º–µ–Ω–µ–Ω–∏—è –¥–∞–Ω–Ω—ã—Ö –º–µ—Ç–æ–¥–æ–≤

### PCA

–°–ø–µ—Ä–≤–∞ –¥–∞–≤–∞–π—Ç–µ –ø–æ—Å—á–∏—Ç–∞–µ–º "—Ä—É–∫–∞–º–∏"

```python
#center columns by subtracting column means
X_centered = X - np.mean(X, axis=0)
# calculate scatter matrix of centered matrix
scatter_matrix = X_centered.T @ X_centered

################################################################################
#or we can compute covariance matrix in a way and use it instead scatter matrix#
#covariance_matrix = np.cov(X_centered.T)                                      #
#np.allclose(covariance_matrix * (len(X_centered) - 1), scatter_matrix)        #
#######################True#####################################################

# eigen decomposition of scatter matrix
eig_val_sc, eig_vec_sc = np.linalg.eig(scatter_matrix)
#project data
projections = X_centered @ eig_vec_sc
```

–¢–µ–ø–µ—Ä—å –ø—Ä–∏ –ø–æ–º–æ—â–∏ sklearn.PCA

```python
%%time
pca = PCA(n_components=2, random_state=SEED).fit(X)
X_reduc = pca.transform(X)
```

```python
flag = True
for i in range(X_reduc.shape[1]):
    #'+' is used due to values have '-' from pca where our have '+' and vice versa
    if sum(projections[:, i] + X_reduc[:, i]) > 1e-10:
        flag = False
        print(f"Something wrong")
if flag:
    print(f"Great, by hand projections looks the same as from sklearn.PCA")
```

```python
plt.figure(figsize=(15, 10))
plt.scatter(X_reduc[:, 0], X_reduc[:, 1], c=y, s=20, cmap=plt.cm.get_cmap('nipy_spectral', 10))
plt.colorbar()
plt.title('MNIST data, PCA projection', fontsize=13);
```

–ö–ª–∞—Å—Å—ã –≤–∏–∑—É–∞–ª—å–Ω–æ –ø–ª–æ—Ö–æ —Ä–∞–∑–¥–µ–ª–∏–º—ã, –º–µ—Å—Ç–∞–º–∏ –ø–µ—Ä–µ–º–µ—à–∞–Ω—ã, –ø–æ–ø—Ä–æ–±—É–µ–º t-sne
